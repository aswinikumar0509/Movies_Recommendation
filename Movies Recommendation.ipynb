```python
import numpy as np
import pandas as pd
```

```python
movies_data = pd.read_csv('movies.csv')
movies_data.head(3)
```

```python
movies_data.size
```

```python
movies_data.shape
```

```python
select_feature = ['genres','keywords','tagline','cast','director']
select_feature
```

```python
# replacing the null value in the selected feature
```

```python
for feature in select_feature:
    movies_data[feature] = movies_data[feature].fillna(' ')
```

```python
combined_feature = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']
```

```python
print(combined_feature)
```

```python
#Converting the text data into  feature vector
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_feature)
```

```python
print(feature_vectors)
```

# getting the similarity scoreds using cosine similarity

Cosine similarity is a measure of similarity, often used to measure document similarity in text analysis

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(feature_vectors)
print(similarity)
```

# Getting the name of the movies from the user

```python
movie_name = input("Enter the name of the movies : ")
```

```python
list_of_all_titles = movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
import difflib
```

```python
close_match = difflib.get_close_matches(movie_name,list_of_all_titles)
print(close_match)
```

```python
match = close_match[0]
print(match)
```

```python
index_of_the_movies = movies_data[movies_data.title == match]['index'].values[0]
```

```python
print(index_of_the_movies)
```

```python
similarity_score = list(enumerate(similarity[index_of_the_movies]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True) 
print(sorted_similar_movies)
```

# Simmilar movies based on the index

```python
print('Recommended movies : \n')

i=1

for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = movies_data[movies_data.index==index]['title'].values[0]
    if(i<30):
        print(i,'.',title_from_index)
        i+=1
```

```python
movie_name = input("Enter the name of the movies : ")

list_of_all_titles = movies_data['title'].tolist()
close_match = difflib.get_close_matches(movie_name,list_of_all_titles)
match = close_match[0]
index_of_the_movies = movies_data[movies_data.title == match]['index'].values[0]
similarity_score = list(enumerate(similarity[index_of_the_movies]))
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True) 

print('Recommended movies : \n')

i=1

for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = movies_data[movies_data.index==index]['title'].values[0]
    if(i<10):
        print(i,'.',title_from_index)
        i+=1
```
